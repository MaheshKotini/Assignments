```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
df = pd.DataFrame(data['data'],columns=data['feature_names'])
```

```python
data.keys()
```

```python
df.head()
```

```python
dec = PCA(n_components=3)
```

```python
dec.fit(df)
```

```python
scaled_data = dec.transform(df)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(scaled_data[:,0],scaled_data[:,1],scaled_data[:,2],c=data['target'])
plt.xlabel("X - axis")
plt.ylabel("y - axis")
```
